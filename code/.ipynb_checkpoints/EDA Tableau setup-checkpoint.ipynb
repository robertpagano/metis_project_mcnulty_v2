```python
cat_kitten = df_X.groupby(['cat_or_kitten', 'outcome_type_Transfer']).agg({'outcome_type_Transfer':'count'})
cat_kitten.columns = ['outcome_by_age']
cat_kitten.reset_index(inplace=True)
cat_kitten
```

```python
cat_kitten_sum = cat_kitten.groupby('cat_or_kitten').agg('sum')
cat_kitten_sum
```

```python
cat_kitten = cat_kitten.merge(cat_kitten_sum, on='cat_or_kitten')
cat_kitten
```

```python
cat_kitten = cat_kitten.rename(columns={'outcome_type_Transfer_x': 'transfer', 'outcome_type_Transfer_y': 'total_age'})
```

```python
cat_kitten = cat_kitten.loc[:, ['cat_or_kitten', 'transfer', 'outcome_by_age', 'total_age']]
cat_kitten
```

```python
cat_kitten['proportion'] = cat_kitten['outcome_by_age'] / cat_kitten['total_age']
cat_kitten
```
